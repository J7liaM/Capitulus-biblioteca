```python
import matplotlib.pyplot as pyplot

class Livro:
    def __init__(self, titulo, autor, genero, quantidade):
        self.titulo = titulo
        self.autor = autor
        self.genero = genero
        self.quantidade = quantidade

# Lista de livros
ArmazenamentoLiv = []

# Fun√ß√£o de cadastrar livro
def cadastrar_livros():
    print("\n Cadastro de Livro ")
    titulo = input("Digite o t√≠tulo do livro: ")
    autor = input("Digite o autor do livro: ")
    genero = input("Digite o g√™nero do livro: ")
    quantidade = int(input("Digite a quantidade do livro: "))
    novo_livro = Livro(titulo, autor, genero, quantidade)
    ArmazenamentoLiv.append(novo_livro)
    print("üìö Livro cadastrado com sucesso!")

# Fun√ß√£o para listar livros
def listar_livros():
    print("\n Lista de Livros ")
    if not ArmazenamentoLiv:
        print("Nenhum livro cadastrado.")
    else:
        for livro in ArmazenamentoLiv:
            print(f"T√≠tulo: {livro.titulo}, Autor: {livro.autor}, G√™nero: {livro.genero}, Quantidade: {livro.quantidade}")

# Fun√ß√£o para buscar livro pelo t√≠tulo
def buscar_livro():
    print("\n Buscar Livro ")
    titulo_busca = input("Digite o t√≠tulo do livro que deseja buscar: ").lower()
    for livro in ArmazenamentoLiv:
        if livro.titulo.lower() == titulo_busca:
            print(f"T√≠tulo: {livro.titulo}, Autor: {livro.autor}, G√™nero: {livro.genero}, Quantidade: {livro.quantidade}")
            return
    print("Livro n√£o encontrado.")

# Fun√ß√£o para gerar gr√°fico
def gerar_grafico():
    if not ArmazenamentoLiv:
        print("Nenhum livro cadastrado. N√£o √© poss√≠vel gerar o gr√°fico.")
        return

    generos = {}
    for livro in ArmazenamentoLiv:
        if livro.genero in generos:
            generos[livro.genero] += livro.quantidade
        else:
            generos[livro.genero] = livro.quantidade

    pyplot.title("Quantidade de livros por g√™nero")
    pyplot.bar(generos.keys(), generos.values())
    pyplot.show()

# Menu principal
while True:
    print("\n Menu capitulus ")
    print("(1) Cadastrar livro")
    print("(2) Listar todos os livros")
    print("(3) Buscar um livro pelo t√≠tulo")
    print("(4) Gerar gr√°fico")
    print("(5) Sair")

    escolha = input("Selecione a fun√ß√£o que deseja: ")

    if escolha == '1':
        cadastrar_livros()
    elif escolha == '2':
        listar_livros()
    elif escolha == '3':
        buscar_livro()
    elif escolha == '4':
        gerar_grafico()
        break
    elif escolha == '5':
        print("Obrigado pela presen√ßa, e boa leitura! üìñ")
        break
    else:
        print("Op√ß√£o inv√°lida.")
```
